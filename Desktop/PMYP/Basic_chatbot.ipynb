## Task 01 (19-03-2025):

Using the Python fundamentals we have learned (variables, data types, operators, input handling, conditionals, loops, functions, and string operations), create a simple chatbot that can:

1. Greet the user and ask for their name.
2. Display a menu with options:
3. Perform basic math operations (addition, subtraction, multiplication, division).
4. Perform string operations (convert to uppercase/lowercase, reverse a string, count words).
4. Check if a person is a minor or an adult based on age input.
5. Exit the chatbot.

Note: Use a loop to keep the chatbot running until the user chooses to exit.
Handle invalid inputs with appropriate messages.
Write a Python program to implement this chatbot and test it by interacting with it.

Bonus: Improve the chatbot by adding new features like time-based greetings, user history, or additional functionalities.

```python
print("Hello! Welcome to this chatbot.")
name = input("What's your name? ")

while True:
    print(name,", please choose an option:")
    print("1. Perform basic math operations")
    print("2. Perform string operations")
    print("3. Check if you are a minor or an adult")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("\nChoose a math operation:")
        print("1. Addition (+)\n2. Subtraction (-)\n3. Multiplication (*)\n4. Division (/)")
        
        operation = input("Enter your choice: ")

        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if operation == "1":
            print("Result:",num1 ,"+", num2, "=", num1 + num2)
        elif operation == "2":
            print("Result:" ,num1,"-",num2,"=", num1 - num2)
        elif operation == "3":
            print("Result:" ,num1,"x",num2,"=", num1*num2)
        elif operation == "4":
            if num2 == 0:
                print("Error: Division by zero is not allowed.")
            else:
                print("Result:" ,num1,"/",num2,"=", num1/num2)
        else:
            print("Invalid choice! Please try again.Choose the correct number")

    elif choice == "2":
        string = input("\nEnter a string: ")

        print("\nChoose a string operation:")
        print("1. Convert to uppercase\n2. Convert to lowercase\n3. Reverse string\n4. Count words")

        string_choice = input("Enter your choice: ")

        if string_choice == "1":
            print("Uppercase String: ",string.upper())
        elif string_choice == "2":
            print("Lowercase String: ",string.lower())
        elif string_choice == "3":
            print("Reversed String: ",string[::-1])
        elif string_choice == "4":
            print("Word Count: ", len(string))
            
        else:
            print("Invalid choice! Please try again.")

    elif choice == "3":
        age = int(input("\nEnter your age: "))
        if age<18:
            print("YOU ARE A MINOR")
        else:
            print("YOU ARE ADULT")

    elif choice == "4":
        print("Goodbye,",name,"! Have a great day! ")
        break

    else:
        print("Invalid choice! Please try again.")
```
